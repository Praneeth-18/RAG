```python
import re
import contractions

def clean_txt(text):
    # Replace specific Unicode characters and whitespace characters with space
    text = text.replace(u'\xa0', u' ').replace('\r', '\n').replace('\n', ' ').replace('\t', ' ')

    # Expand contractions, e.g., "don't" to "do not"
    text = contractions.fix(text)

    # Strip leading/trailing whitespace and convert to lowercase
    text = text.strip().lower()

    # Replace apostrophes with space
    text = re.sub(r"'", ' ', text)

    # Define characters to filter out
    filters = '!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n'
    translate_dict = dict((c, " ") for c in filters)  # Map each punctuation to space
    translate_map = str.maketrans(translate_dict)
    text = text.translate(translate_map)

    # Remove single-character words and extra spaces
    text = ' '.join([w for w in text.split() if len(w) > 1])
    text = re.sub(' +', ' ', text)

    return text
```

```python
def scrapePoemFromURL(poet, poem_url_list):
    poems_results = []
    for url in poem_url_list:
        header = {'User-Agent': 'Mozilla/5.0'}
        req = urllib.request.Request(url, headers=header)
        request_read = urllib.request.urlopen(req).read()
        soup = bs.BeautifulSoup(request_read, 'html.parser')

        poem_for_human = soup.find_all('div', class_="poem_body")[0].text.replace("© by owner. provided at no charge for educational purposes", "")
        poem_for_machine = clean_txt(poem_for_human)
        poem_title = soup.find_all('h1')[0].text

        poem_document = {
            "poet": poet,
            "title": poem_title,
            "poem_for_machine": poem_for_machine,
            "poem_for_human": poem_for_human,
            "url": url
        }

        poems_results.append(poem_document)

    poems_collection.insert_many(poems_results)
    return poems_results
```

```python
def get_poem_url_json(poet):
    site = "https://allpoetry.com"
    search = "https://allpoetry.com/items/read_by/"
    page = 1
    URL_list = []
    while len(URL_list) < 10:
        url = search + poet + "?page=" + str(page)
        response = requests.get(url)
        soup = bs.BeautifulSoup(response.text, 'lxml')
        poem_list = soup.find_all('a', href=True, class_='nocolor fn')
        for poem in poem_list:
            URL_list.append(site + poem['href'])
            if len(URL_list) == 10:
                break
        page += 1
    return {"poet": poet, "URL": URL_list}
```

```python
# Embedding setup
model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to insert a poem and its embedding
def insert_poem(poem_text, poem_title):
    poem_embedding = model.encode(poem_text)
    poem_document = {
        "title": poem_title,
        "text": poem_text,
        "embedding": poem_embedding.tolist()  # Store list of floats
    }
    poems_collection.insert_one(poem_document)

# Insert the two poems into MongoDB
insert_poem(tse_morning_poem, "TSE Morning Poem")
insert_poem(rk_morning_poem, "RK Morning Poem")

# Function to retrieve a poem's vector from MongoDB
def get_poem_vector(title):
    poem = poems_collection.find_one({"title": title})
    return np.array(poem['embedding'])

# Retrieve vectors for both poems
vec_tse_morning = get_poem_vector("TSE Morning Poem")
vec_rk_morning = get_poem_vector("RK Morning Poem")

# Cosine similarity function
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# Compute similarity
similarity_score = cosine_similarity(vec_tse_morning, vec_rk_morning)
print("Cosine Similarity between TSE Morning Poem and RK Morning Poem:", similarity_score)
```

```python
from sentence_transformers import SentenceTransformer

# Initialize the Sentence Transformer Model
model = SentenceTransformer('all-MiniLM-L6-v2')

def embed_poem_lines(poem):
    """Embed each line of the poem using a transformer model."""
    lines = poem.split('\n')  # Split poem into lines
    embeddings = model.encode(lines)  # Encode each line
    return lines, embeddings.tolist()  # Convert numpy array to list for MongoDB storage

def store_line_embeddings(title, poet):
    """Fetch the poem, embed each line, and store the embeddings back to MongoDB."""
    poem_document = poems_collection.find_one({"title": title, "poet": poet})
    if poem_document:
        lines, line_embeddings = embed_poem_lines(poem_document['poem_for_human'])
        # Update the document with line embeddings
        poems_collection.update_one(
            {"_id": poem_document['_id']},
            {"$set": {"line_embeddings": line_embeddings}}
        )
        print(f"Embeddings stored for {title} by {poet}")

# Example usage: Store line embeddings for "If...." by Rudyard Kipling
store_line_embeddings("If....", "Rudyard Kipling")
```

```python
def print_poem_embeddings(title, poet):
    """Retrieve and print the line embeddings for a specific poem from MongoDB."""
    poem_document = poems_collection.find_one({"title": title, "poet": poet})
    if poem_document and "line_embeddings" in poem_document:
        lines = poem_document['poem_for_human'].split('\n')  # Re-split the poem into lines
        line_embeddings = poem_document['line_embeddings']

        print(f"Line embeddings for '{title}' by {poet}:")
        for line, embedding in zip(lines, line_embeddings):
            print(f"Line: {line}")
            print("Embedding:", embedding)
    else:
        print(f"No embeddings found for '{title}' by {poet}.")

# Example usage: Retrieve and print line embeddings for "If...." by Rudyard Kipling
print_poem_embeddings("If....", "Rudyard Kipling")
```
